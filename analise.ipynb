```python
import pandas as pd

def load_genome():
    dataset = []
    filepath = "ncbi_dataset/data/genomic.fna"
    with open(filepath) as file:
        total_lines = 0
        line = file.readline()
        while line:
            line = line.strip()
            if ("complete genome" not in line) or ("SARS-CoV-2" not in line):
                line = file.readline()
                continue
            total_lines += 1
            description = line
#             print(":::",line)
            
            
            
            content = ""
            while True:
                line = file.readline()
                if not line:
                    break
                if line[0] == ">":
                    break
                content += line.strip()
            
            if total_lines > 100:
                break

#             print(total_lines, len(content))
            dataset.append([total_lines, len(content), description, content])
    return pd.DataFrame(dataset, columns=["sample", "length", "description", "content"])
```

```python
dataset = load_genome()
dataset.describe()
```

```python
def diferenca(s1, s2):
    total = 0
    diff = ""
    for (position, (c1, c2)) in enumerate(zip(s1, s2)):
        if c1 != c2:
            total += 1
            diff += f"{position} {c1}{c2}|"
    return total, diff[:500]
```

```python
s1 = dataset.iloc[index1]["content"]
s2 = dataset.iloc[index2]["content"]
diferenca(s1, s2)
```

```python
def compare_all(dataset, length):
    subset = dataset.query(f"length=={length}")
    first = subset.iloc[0]
    for other_index, other in subset[1:].iterrows():
        total, diff = diferenca(first['content'], other['content'])
        print(total, diff)
```
